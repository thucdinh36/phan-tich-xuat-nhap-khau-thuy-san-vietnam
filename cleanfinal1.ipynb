```python
import pandas as pd

# 1. Đọc dữ liệu từ file CSV ban đầu
file_path = 'UN_VNM_Seafood_Export_ByPartner_HS03_1604_1605_2015_2023.csv'
df = pd.read_csv(file_path, encoding="latin1", index_col=False)

# 2. Định nghĩa các cột cần giữ lại và tên tiếng Anh rút gọn
columns_mapping = {
    'refYear': 'Year',
    'flowDesc': 'Trade_Flow',
    'partnerISO': 'Partner_ISO',
    'partnerDesc': 'Partner_Country',
    'cmdCode': 'HS_Code',
    'cmdDesc': 'Product_Description',
    'primaryValue': 'Value_USD'
}

# 3. Lọc lấy các cột chính và đổi tên cột
df_filtered = df[list(columns_mapping.keys())].rename(columns=columns_mapping)

# 4. Loại bỏ dòng tổng toàn thế giới "World" để tránh trùng lặp dữ liệu các quốc gia
df_filtered = df_filtered[df_filtered['Partner_Country'] != 'World']

# 5. Lưu kết quả ra file mới gọn gàng hơn
output_path = 'VNM_Seafood_Export_ByPartner_Cleaned_EN.csv'
df_filtered.to_csv(output_path, index=False, encoding='utf-8')

print("Đã lọc và đổi tên thành công! File mới lưu tại:", output_path)
print("\nXem trước dữ liệu (Preview):")
print(df_filtered.head())
```
