```python
import pandas as pd

# 1. Đọc dữ liệu từ file CSV gốc (Tổng quan Xuất Nhập Khẩu)
file_path_2 = 'UN_VNM_Seafood_World_ExportImport_HS03_1604_1605_2015_2023.csv'
df2 = pd.read_csv(file_path_2, encoding="latin1", index_col=False)

# 2. Định nghĩa các cột cần giữ lại (Đã bổ sung cột khối lượng netWgt)
columns_mapping_2 = {
    'refYear': 'Year',
    'flowDesc': 'Trade_Flow',
    'partnerDesc': 'Partner_Country',
    'cmdCode': 'HS_Code',
    'cmdDesc': 'Product_Description',
    'netWgt': 'Weight_KG',              # Cột khối lượng thực tế (kg)
    'primaryValue': 'Value_USD'
}

# 3. Lọc lấy các cột chính và đổi tên cột
df2_filtered = df2[list(columns_mapping_2.keys())].rename(columns=columns_mapping_2)

# 4. [Tùy chọn nâng cao] Bạn có thể điền giá trị 0 cho những dòng bị trống khối lượng
df2_filtered['Weight_KG'] = df2_filtered['Weight_KG'].fillna(0)

# 5. Lưu kết quả ra file mới gọn gàng hơn
output_path_2 = 'VNM_Seafood_World_Trade_With_Weight.csv'
df2_filtered.to_csv(output_path_2, index=False, encoding='utf-8')

print("Đã lọc dữ liệu thành công! Lưu tại:", output_path_2)
print("\nXem trước dữ liệu (Preview):")
print(df2_filtered.head())
```
