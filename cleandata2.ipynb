```python
import pandas as pd

# Đọc file CSV
file_path = "UN_VNM_Seafood_World_ExportImport_HS03_1604_1605_2015_2023.csv"

df = pd.read_csv(file_path)

# Các cột cần giữ
columns_keep = [
    "refYear",
    "refMonth",
    "reporterDesc",
    "flowDesc",
    "partnerDesc",
    "cmdCode",
    "cmdDesc",
    "qty",
    "qtyUnitAbbr",
    "netWgt",
    "fobvalue",
    "cifvalue",
    "primaryValue"
]

# Giữ lại các cột
df_filtered = df[columns_keep]

# Xuất file mới
output_file = "Seafood_Filtered.csv"
df_filtered.to_csv(output_file, index=False, encoding="utf-8-sig")

print("Đã xuất file:", output_file)
print(df_filtered.head())
```
