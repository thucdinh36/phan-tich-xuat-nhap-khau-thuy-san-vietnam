```python
import pandas as pd

# 1. Đọc dữ liệu file 2024-2025
file_path_3 = 'VN_Seafood_2024_2025_Customs_VITIC.csv'
df3 = pd.read_csv(file_path_3)

# 2. Lọc lấy các cột cần thiết và đổi tên cho gọn
columns_mapping_3 = {
    'Year': 'Year',
    'Flow': 'Trade_Flow',
    'Partner': 'Partner_Country',
    'Trade_Value_USD': 'Value_USD'
}
df3_filtered = df3[list(columns_mapping_3.keys())].rename(columns=columns_mapping_3)

# 3. Đồng bộ hóa giá trị chuỗi (ví dụ: 'Exports' -> 'Export' cho khớp file cũ)
df3_filtered['Trade_Flow'] = df3_filtered['Trade_Flow'].replace({'Exports': 'Export', 'Imports': 'Import'})

# 4. TỰ ĐỘNG TẠO THÊM CỘT CODE và các cột thiếu để đồng bộ cấu trúc
df3_filtered['HS_Code'] = 'TOTAL'  # Tạo cột mã sản phẩm chung
df3_filtered['Product_Description'] = 'Total Seafood'
df3_filtered['Weight_KG'] = 0.0     # Điền 0 vì dữ liệu gốc không có khối lượng
df3_filtered['Partner_ISO'] = ''     # Để trống hoặc điền mã tùy ý

# 5. Sắp xếp lại thứ tự các cột cho khớp hoàn toàn với File cũ
desired_columns = ['Year', 'Trade_Flow', 'Partner_Country', 'HS_Code', 'Product_Description', 'Weight_KG', 'Value_USD']
df3_final = df3_filtered[desired_columns]

# 6. Lưu ra file mới sạch sẽ
output_path_3 = 'VNM_Seafood_2024_2025_Cleaned_EN.csv'
df3_final.to_csv(output_path_3, index=False, encoding='utf-8')

print("Đã chuẩn hóa và tạo cột Code thành công! Lưu tại:", output_path_3)
print("\nXem trước dữ liệu sau khi đồng bộ cột:")
print(df3_final.head())
```
